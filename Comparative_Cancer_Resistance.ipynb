```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Species": [
        "Human",
        "Elephant",
        "Bowhead Whale",
        "Tardigrade"
    ],

    "Lifespan_Years": [
        80,
        70,
        200,
        1
    ]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    df["Species"],
    df["Lifespan_Years"]
)

plt.title("Lifespan Across Species")

plt.ylabel("Years")

plt.show()
```

# Observation

Large and long-lived organisms do not always show higher cancer incidence.

This phenomenon is known as Peto's Paradox.

Elephants evolved multiple TP53 copies.

Whales appear to possess enhanced DNA maintenance systems.

Tardigrades possess Dsup protein which protects DNA against radiation damage.

```python
species = [
    "Human",
    "Elephant"
]

tp53 = [
    2,
    40
]

plt.figure(figsize=(6,4))

plt.bar(
    species,
    tp53
)

plt.title("Approximate TP53 Copy Numbers")

plt.ylabel("Gene Copies")

plt.show()
```
